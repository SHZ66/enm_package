```python
from pyrwr.pagerank import PageRank

pagerank = PageRank()
pagerank.read_graph('../data/interim/costanzo_pcc_ALL_single_weighted.tsv','undirected')
r = pagerank.compute(.5)
```

```python
rwr = RWR()
rwr.read_graph('../data/interim/costanzo_pcc_ALL_single_weighted.tsv','undirected')
```

```python
rwr.A.shape
```

```python
rwr_mat = np.zeros(rwr.A.shape)
```

```python
for i in range(rwr_mat.shape[0]):
    seed = i 
    r = rwr.compute(seed, c = 0.5)
    rwr_mat[i,:] = r
    
```

```python
sens = np.mean(rwr_mat,axis=0)
```

```python
plt.plot(sens)
```

```python
plt.plot(r,sens)
```
